```python
transpose_image5 = image1.transpose(Image.FLIP_LEFT_RIGHT)
transpose_image6 = image1.transpose(Image.FLIP_TOP_BOTTOM)
transpose_image7 = image1.transpose(Image.ROTATE_90)
transpose_image8 = image1.transpose(Image.ROTATE_180)
transpose_image9 = image1.transpose(Image.ROTATE_270)
transpose_image10 = image1.transpose(Image.TRANSPOSE)

titles = ['original image','FLIP_RIGHT', 'FLIP_TOP' ,'ROTATE 90', "ROTATE 180", 'ROTATE 270', 'TRANSPOSE']
images = [image1,transpose_image5,transpose_image6,transpose_image7,transpose_image8,transpose_image9,transpose_image10]

for i in range(7):
    plt.subplot(4,3, i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])

plt.show()
```

```python
newsize = (300,300)
plt.figure(figsize=(10,10))
resize1= image1.resize(newsize,Image.BILINEAR)
resize2= image1.resize(newsize,Image.NEAREST)
resize3= image1.resize(newsize,Image.BOX)
resize4= image1.resize(newsize,Image.HAMMING)
resize5= image1.resize(newsize,Image.BICUBIC)
resize6= image1.resize(newsize,Image.LANCZOS)

titles = ['original image','BILNEAR', 'NEAREST' ,'BOX', "HAMMING", 'BICUBIC', 'LANCZOS']
images = [image1,resize1,resize2,resize3,resize4,resize5,resize6]

for i in range(7):
    plt.subplot(4,3, i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])

plt.show()
```

```python
from PIL import ImageEnhance
plt.figure(figsize=(10,10))
image_color_enhace = image1.copy()
enhance1 = ImageEnhance.Color(image_color_enhace).enhance(2.5)
enhance2 = ImageEnhance.Contrast(image_color_enhace).enhance(2.5)
enhance3 = ImageEnhance.Brightness(image_color_enhace).enhance(1.5)
enhance4 = ImageEnhance.Sharpness(image_color_enhace).enhance(2.5)

titles = ['original image','COLOR', 'CONTRAST' ,'BRIGHTNESS', "SHARPNESS"]
images = [image1,enhance1,enhance2,enhance3,enhance4]

for i in range(5):
    plt.subplot(2,3, i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])

plt.show()
```
